```python
#Import required libraries from qiskit
from qiskit import QuantumCircuit
from qiskit_aer import Aer
from qiskit_aer import AerSimulator 
from qiskit.visualization import plot_histogram, plot_bloch_multivector
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
#Create quantum circuit with one qubit and one classical bit
qc = QuantumCircuit(1, 1)
#apply Hadamard gate to create superposition
qc.h(0)
#apply Z gate 
qc.z(0) 
#measure the qubit 
qc.measure(0, 0)
#draw circuit 
print("quantum circuit")
fig, ax = plt.subplots()  
qc.draw('mpl', ax=ax)
plt.show() 
```

```python
#simulate the circuit using AerSimulator
simulator = AerSimulator(method='automatic')
result = simulator.run(qc, shots=1024).result()
#measurement results
counts = result.get_counts(qc)
#histogram of results
print("Measurement Results:")
fig, ax = plt.subplots() 
plot_histogram(counts, ax=ax)
plt.show()
```

```python
#create a separate circuit for statevector analysis (no measurement)
qc_state = QuantumCircuit(1)
qc_state.h(0)
qc_state.z(0)

#Explicitly save statevector in the circuit
qc_state.save_statevector() 

# Simulate statevector using AerSimulator with statevector method
simulator_state = AerSimulator(method='statevector')
result_state = simulator_state.run(qc_state).result()

#Get statevector
statevector = result_state.get_statevector(qc_state)
```

```python
#Plot Bloch sphere
print("Bloch Sphere for Qubit State:")
display(plot_bloch_multivector(statevector))
```
